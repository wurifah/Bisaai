```python
# import library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#pembacaan dataset

aktivitas = pd.read_excel('perawatan.xlsx')
aktivitas
```

# Cleaning dan Preprocessing

Agar pembacaan dataset lebih mudah, maka kita ganti semua nama kolomnya sehingga semua hurufnya menjadi huruf kecil dan spasi menjadi underscore (_)

```python
aktivitas.columns = aktivitas.columns.str.replace(' ','_').str.lower()
aktivitas
```

```python
aktivitas.isnull().sum()
```

Dataset terdapat banyak missing value. Kebanyakan missing value sejumlah 491. Maka dari itu kita drop 491 missing value tersebut. Karena kita akan menganalisis dataset berdasarkan jenis_layanan, maka kita drop missing value dengan berpacu pada kolom jenis_layanan

```python
aktivitas.dropna(subset=["jenis_layanan"], axis=0, inplace=True)
```

```python
aktivitas.isnull().sum()
```

# Analisis

Analisis dilakukan dengan beberapa tujuan yaitu:

1. Mengetahui jenis layanan terbanyak

2. Mengetahui frekuensi jenis layanan error berdasarkan lokasi

3. Mengetahui frekuensi jenis layanan error berdasarkan bulan

## 1. Jenis layanan terbanyak

```python
lokasi = aktivitas['jenis_layanan'].value_counts()
print(lokasi)
```

```python
plt.figure(figsize=(12,8))
sns.set_theme(style="darkgrid")
small_to_large = aktivitas.groupby('jenis_layanan').size().sort_values().index[::1]
ax = sns.countplot(x='jenis_layanan',data=aktivitas, order=small_to_large)
plt.title('Grafik Frekuensi Jenis Layanan', fontsize = 20)
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+0))
plt.ylabel('Frekuensi', fontsize = 15)
plt.xlabel('Jenis Layanan', fontsize = 15)
plt.xticks(rotation=90)
plt.savefig('layanan.png')
plt.show()
```

## 2. Frekuensi jenis layanan error berdasarkan lokasi

Lalu kita akan membahas lebih lanjut Jenis Layanan yang jenisnya error. Kita satukan jenis layanan yang jenisnya error dengan error & PM. Agar lebih mudah saat pembacaan data, kita jadikan data pada kolom jenis_layanan menjadi huruf kecil

```python
aktivitas['jenis_layanan_kecil'] = aktivitas['jenis_layanan'].str.lower()
aktivitas.shape
```

```python
error = aktivitas[aktivitas['jenis_layanan_kecil'].str.contains('error')]
error
```

```python

plt.figure(figsize=(10,60))
sns.set_theme(style="darkgrid")
small_to_large = error.groupby('lokasi').size().sort_values().index[::1]
ax = sns.countplot(y='lokasi',data=error,palette='winter', order=small_to_large)
plt.title('Grafik Frekuensi Error Berdasarkan Lokasi', fontsize = 25)
plt.ylabel('Lokasi', fontsize = 20)
plt.xlabel('Frekuensi', fontsize = 20)
plt.savefig('a_frekuensi_lokasi.png')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.bar(df.bulan,df.jumlah, color = 'blue')
plt.title('Grafik Jumlah Error per-Bulan', fontsize = 20)
for rect in ax:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2., 1.0*height,'%.0f' % float(height),ha='center', va='bottom')
plt.ylabel('Jumlah Error per-Bulan', fontsize = 15)
plt.xlabel('Bulan', fontsize = 15)
plt.xticks(rotation=90)
plt.savefig('bulan.png')
plt.show()
```
