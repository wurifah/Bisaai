```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data covid jakarta maret-mei.csv')
df
```

```python
# seleksi data positif
positif = df[df['diagnosa'] =='Positif Harian']
positif
```

```python
#plotting data positif
positif.plot(x='tanggal',y='jumlah', kind = 'bar', color =  'blue',figsize=(20,10))
plt.show()
```

```python
#seleksi data meninggal
meninggal = df[df['diagnosa'] =='Meninggal Harian']
meninggal
```

```python
#plotting data meninggal
meninggal.plot(x='tanggal',y='jumlah', kind = 'bar', color = 'red',figsize=(20,10))
plt.show()
```

```python
#seleksi data sembuh
sembuh = df[df['diagnosa'] =='Sembuh Harian']
sembuh
```

```python
#plotting data sembuh
sembuh.plot(x='tanggal',y='jumlah', kind = 'bar', color = 'green',figsize=(20,10))
plt.show()
```

```python
#plotting data positif, meninggal dan sembuh
plt.figure(figsize=(20,10))
plt.bar(positif.tanggal,positif.jumlah, color = 'blue')
plt.bar(meninggal.tanggal,meninggal.jumlah, color = 'red')
plt.bar(sembuh.tanggal,sembuh.jumlah, color = 'green')
plt.xticks(rotation=90)
plt.show()
```

```python
#menghitung jumlah kumulatif
df['jumlahkumulatif'] = df.groupby('diagnosa')['jumlah'].cumsum()
df
```

```python
#jumlah kumulatif positif
positif = df[df['diagnosa'] =='Positif Harian']
positif
```

```python
#plotting jumlah kumulatif positif
positif.plot(x='tanggal',y='jumlahkumulatif', kind = 'line',figsize=(10,5))
plt.show()
```

```python
#jumlah kumulatif meninggal
meninggal = df[df['diagnosa'] =='Meninggal Harian']
meninggal
```

```python
#plotting jumlah kumulatif meninggal
meninggal.plot(x='tanggal',y='jumlahkumulatif', kind = 'line',color = 'red',figsize=(10,5))
plt.show()
```

```python
#jumlah kumulatif sembuh
sembuh = df[df['diagnosa'] =='Sembuh Harian']
sembuh
```

```python
#plotting jumlah kumulatif sembuh
sembuh.plot(x='tanggal',y='jumlahkumulatif', kind = 'line',color = 'green',figsize=(10,5))
plt.show()
```

```python
#plotting jumlah kumulatif keseluruhan
plt.figure(figsize=(20,10))
plt.plot(positif.tanggal,positif.jumlahkumulatif, color = 'blue')
plt.plot(meninggal.tanggal,meninggal.jumlahkumulatif, color = 'red')
plt.plot(sembuh.tanggal,sembuh.jumlahkumulatif, color = 'green')
plt.legend(['Positif','Meninggal','Sembuh'])
plt.xticks(rotation=90)
plt.show()
```
