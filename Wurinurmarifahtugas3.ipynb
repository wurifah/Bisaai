```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
sns.set()
df.plot(x='SepalLengthCm',y='SepalWidthCm', kind = 'scatter',marker = '^')
plt.title('Data Sepal Length dan Sepal Width')

df.plot(x='PetalLengthCm',y='PetalWidthCm', kind = 'scatter',marker = '^')
plt.title('Data Petal Length dan Petal Width')
```

```python
setosa = df[df['Species'] =='Iris-setosa']
setosa
```

```python
setosa.plot(x='SepalLengthCm',y='SepalWidthCm', kind = 'scatter',marker = '^')
plt.title('Data Sepal Length dan Sepal Width Spesies Iris-Setosa')
setosa.plot(x='PetalLengthCm',y='PetalWidthCm', kind = 'scatter',marker = 'o', color='magenta')
plt.title('Data Petal Length dan Petal Width Spesies Iris-setosa')
```

```python
versicolor = df[df['Species'] =='Iris-versicolor']
versicolor
```

```python
versicolor.plot(x='SepalLengthCm',y='SepalWidthCm', kind = 'scatter',marker = '^', color='green')
plt.title('Data Sepal Length dan Sepal Width Spesies Iris-Setosa')
versicolor.plot(x='PetalLengthCm',y='PetalWidthCm', kind = 'scatter',marker = 'o', color='red')
plt.title('Data Petal Length dan Petal Width Spesies Iris-setosa')
```

```python
virginica = df[df['Species'] =='Iris-virginica']
virginica
```

```python
virginica.plot(x='SepalLengthCm',y='SepalWidthCm', kind = 'scatter',marker = '^', color='black')
plt.title('Data Sepal Length dan Sepal Width Spesies Iris-Setosa')
virginica.plot(x='PetalLengthCm',y='PetalWidthCm', kind = 'scatter',marker = 'o', color='orange')
plt.title('Data Petal Length dan Petal Width Spesies Iris-setosa')
```

```python
plt.figure(figsize=(20,10))
plt.scatter(setosa.SepalLengthCm , setosa.SepalWidthCm, marker = 'x', color = 'purple')
plt.scatter(virginica.SepalLengthCm , virginica.SepalWidthCm, marker= '^')
plt.scatter(versicolor.SepalLengthCm , versicolor.SepalWidthCm)
plt.legend(['Iris setosa','Iris virginica','Iris versicolor'])
plt.xlabel('Sepal length (cm)',size = 20)
plt.ylabel('Sepal width (cm)',size = 20)
plt.title('Grafik perbandingan Sepal length dan Sepal width',size = 20)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(setosa.PetalLengthCm , setosa.PetalWidthCm, marker = 'x', color = 'black')
plt.scatter(virginica.PetalLengthCm , virginica.PetalWidthCm, marker= '^', color = 'magenta')
plt.scatter(versicolor.PetalLengthCm , versicolor.PetalWidthCm, color = 'blue')
plt.legend(['Iris setosa','Iris virginica','Iris versicolor'])
plt.xlabel('Petal length (cm)',size = 20)
plt.ylabel('Petal width (cm)',size = 20)
plt.title('Grafik perbandingan Petal length dan Petal width',size = 20)
```
